```python
import pandas as pd
import numpy as np

df=pd.read_csv("/home/atri/ML_practice/student-mat.csv",sep=';')

print(df.head())
```

```python
df2=df[['studytime','failures', 'paid','absences','G1','G2']]
df2.head()
```

```python
d={'yes':1,'no':0}
#df2['famsup']=df2['famsup'].map(d)
df2['paid']=df2['paid'].map(d)
#df2['schoolsup']=df2['schoolsup'].map(d)

df2.head()
```

```python
df.describe()
```

```python
features=np.array(df2)
labels=np.array(df['G3'])

from sklearn import preprocessing

sc=preprocessing.StandardScaler()
scaled_feature=sc.fit_transform(features)

print(labels)
```

```python
from sklearn.model_selection import train_test_split
import pickle
from sklearn import linear_model

```

```python

x_train,x_test,y_train,y_test=train_test_split(scaled_feature,labels,test_size=0.1,random_state=0)
best=0
```

```python
for _ in range(30):
    x_train,x_test,y_train,y_test=train_test_split(scaled_feature,labels,test_size=0.1,random_state=1)
    liner=linear_model.LinearRegression()
    liner.fit(x_train,y_train)
    acc=liner.score(x_test,y_test)
    print(acc)
    if acc>best:
        with open("student_marks.pickle","wb") as f:
            pickle.dump(liner,f)
        best=acc    
    
pickle_op=open("student_marks.pickle","rb")
liner=pickle.load(pickle_op)

```

```python
predic=liner.predict(x_test)
#len(y_test)
#print(predic)
#print(y_test)
for i in range(len(predic)):
    print(round(predic[i]),sc.inverse_transform(x_test[i]),y_test[i])
    
```
